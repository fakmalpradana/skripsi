```python
label_folder = "project-4-at-2022-12-24-16-15-40945b8e/"
bangunan_masks = np.zeros((3304, 4365, 3))
jalan_masks = np.zeros((3304, 4365, 3))
sawah_masks = np.zeros((3304, 4365, 3))
sungai_masks = np.zeros((3304, 4365, 3))
pepohonan_masks = np.zeros((3304, 4365, 3))
rth_masks = np.zeros((3304, 4365, 3))
```

```python
for filename in os.listdir(label_folder):
    if 'bangunan' in filename:
        print(filename)
        bangunan_mask = cv2.imread(label_folder + filename)
        bangunan_mask = np.where(bangunan_mask>0, 101, bangunan_mask)
        bangunan_masks = bangunan_masks + bangunan_mask
    elif 'jalan' in filename:
        print(filename)
        jalan_mask = cv2.imread(label_folder + filename)
        jalan_mask = np.where(jalan_mask>0, 203, jalan_mask)
        jalan_masks = jalan_masks + jalan_mask
    elif 'sawah' in filename:
        print(filename)
        sawah_mask = cv2.imread(label_folder + filename)
        sawah_mask = np.where(sawah_mask>0, 305, sawah_mask)
        sawah_masks =  sawah_masks + sawah_mask
    elif 'sungai' in filename:
        print(filename)
        sungai_mask = cv2.imread(label_folder + filename)
        sungai_mask = np.where(sungai_mask>0, 407, sungai_mask)
        sungai_masks = sungai_masks + sungai_mask
    elif 'pepohonan' in filename:
        print(filename)
        pepohonan_mask = cv2.imread(label_folder + filename)
        pepohonan_mask = np.where(pepohonan_mask>0, 509, pepohonan_mask)
        pepohonan_masks = pepohonan_masks + pepohonan_mask
    elif 'rth' in filename:
        print(filename)
        rth_mask = cv2.imread(label_folder + filename)
        rth_mask = np.where(rth_mask>0, 611, rth_mask)
        rth_masks = rth_masks + rth_mask
```
